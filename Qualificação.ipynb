### Resumo


A demanda crescente e contínua por maior taxa de dados exige um esforço muito grande para  que o gerenciamento de recursos rádio seja feito de forma eficiente. As redes emergentes, por exemplo o 5G, promete maior taxa de dados, menor  latência e alta densidade de dispositivos conectados.
No entanto, por mais promissora que seja a solução , ainda nos deparamos em uma limitação das redes celulares anteriores que é a interferência entre células, a qual pode ser subdivida em intra-celular e inter-celular.
A ICI, irá degradar o desempenho do sistema como um todo, mas principalmente, os usuários de borda de célula são os que mais sofrem com o fenômeno.
Diante desse cenário, faz-se cada vez mais necessário o uso de técnicas que otimize de forma automática o gereciamento dos recursos de rádio em cenários dinâmicos, uma vez que as mesmas não se adaptam de forma eficiente aos cenários com tráfego variável, buscando alcançar uma taxa de transmissão adequada e mitigação de ICI. 
O presente trabalho propõe utilizar técnicas de aprendizado de máquina na otimização dos parametros dos esquemas de ICIC que utilizam reuso fracionário de frequencia aplicados em redes móveis LTE e busca reduzir o impacto da interferência e melhorar o desempenho
Além disso, da analisar dos resultados obtidos da aplicação do algoritmos de nossa abordagem, iremos compara-lo com os resultados obtidos de outro algoritmo de apredizado de máquina, Q-Learning, que foi utilizado em [6] .

# Capítulo 1 -  Introdução

A quantidade de equipamento de usuários, do inglês User Equipament (UE), e os serviços ofertados pelas novas gerações de redes móveis exigem cada vez mais largura de banda. 
De acordo com o reporte da Cisco, o crescimento da população com acesso à internet, de 2018 até 2023, será de 1,4 bilhões, ou seja,  entre saindo de 3,9 bilhões (51% da população global) em 2018 para 5,3 bilhões de usuários totais da Internet (66% da população global) até 2023.[9]
De todas as gerações, o 4G é predominante e supera as demais geraçoes em relação a conectividade, espera-se que até 2023 as conexões 4G serão 46% do total de conexões móveis.[9]

**(Fonte: Cisco Systems Inc, 2020)**

Então, para que o padrão LTE, do inglês Long-Term Evolution, e do LTE-A, do inglês Long-Term Evolution – Advanced, possa atender de forma efeciente à demanda, faz-se necessário realizar um gerenciamento de recursos de rádio [3] de modo que venha a melhorar suas funções de otimização.Esse gerenciamento torna-se mais complexo quando nos deparamos com cenários dinâmicos, nos quais surgem picos de  tráfego inesperados devido uma alta concentração de usuários decorrentes de eventos aleatórios, por exemplo eventos, feiras, jogos, dentre outros, ou seja HotSpots.
Nesse contexto, surgem as abordagens  das  técnicas de coordenação de interferência entre células, do inglês  Inter-cell Interference Coordination (ICIC), que buscam por meio de esquemas de reúso de frequência, do inglês Frequency Reuse (FR), controle de potência, entre outros,  mitigar a interferência ocasionadas por células adjacentes.
Em sistemas LTE, como forma de obter maior capacidade, é utilizado o fator de reúso de frequência 1, mas o uso do mesmo,leva a uma redução significativa  no throughput máximo alcançável[5].
Quando nos referimos à interferência entre células, do inglês Inter-Cell Interference (ICI), classificamos elas em dois tipos que são a interferência intra-celular, ocorre entre os co-canais dentro da mesma célula devido o uso de mesma frequência e o nível potência de transmissão, e a interferência inter-celular que ocorre quando o mesmo canal é usado em células diferentes adjacentes.[10]
Além dos percalços descritos acima, ainda existe a possibilidade do surgimento de demandas inesperadas por tráfego, as quais são decorrentes das densas aglomerações de usuários durante eventos musicais, jogos, feiras, confraternizações e etc.
Essas aglomerações inesperadas de usuários são nomeadas como HotSpots e seu surgimento em cenários de sistemas estáticos irá agravar a ICI além de  resuzir a SINR (do inglês, Signal-to-interference-plus-noise Ratio).





## 1.1 - Escopo e Objetivos do Trabalho

Diante das situações descritas acima, a reutilização de frequência é inevitável, desde que o procedimento seja bem projetado, irá ajudar na mitigação da interferência entre células.
Então, faz-se o uso de técnicas de coordenação de interferência entre células (ICIC) focado nos  esquemas  de reúso fracionário de frequência (FFR), do inglês Fractional Frequency Reuse,com fator diferente de 1,  como  forma de tirar o máximo proveito dos recursos de rádio e mitigar os efeitos da ICI. 
Entretanto, a aplicação das técnicas ICIC de forma justa e proporcional tornam-se mais complexas ao depararmos com com cenários dinâmicos com presença de hotspots.
O presente trabalho visa atacar os problemas citados fazendo uso do algoritmo MAB na configuração dos parâmetros das técnicas ICIC em ambientes dinâmicos, de modo que alcance um nível de eficiência na redução dos efeitos da interferência co-canal e alcance melhores SINR's.
Além disso, efetuar uma análise, comparação do resultados simulados e complexidade de aplicação entre  o algoritmo MAB e a técnica Q-Learning que foi empregada no trabalho [6] que foi tomado como referência para o presente.


Diante disso, o trabalho discutirá as seguintes perguntas

 - Qual a estratégia de aprendizado de máquina mais apropriada para definir dinamicamente os parâmetros do ICIC?
 - Qual a melhor técnica de FFR no cenário com hotspots?
 - Qual estratégia de ICIC é mais favorável para o ajuste de parâmetros?
 - Qual a estrátegia de machine learnig mais indicada ao cenário dinâmico
 - Qual os resultados obtidos ao aplicar o algoritmo de aprendizado de máquina MAB considerando um cenário estático e dinâmico?
 - Qual o desempenho da vazão com  e sem o uso do MAB ?
 - Qual o ganho de SINR com e sem o uso do MAB ?
 - Qual dos algoritmos de aprendizado de máquina apresenta melhores resultados quando aplicado em cenário dinâmico levando em consideração a relação otimização de recurso e complexidade de implantação/simulação?
 - Qual o ganho de desempenho comparado  entre o algoritmo Q-Learning e o MAB?
 - Qual algoritmo apresenta melhor resultado  Q-Learning ou MAB quando aplicado em cenários com alta densidade de usuários?
 - Qual algoritmo apresenta melhor resultado  Q-Learning ou MAB quando aplicado em cenários com demandas inesperadas de recursos de rádio?
 - Ao aplicar o algoritmo de aprendizado de máquina selecionado na parametrização do sistema, qual o desempenho obtido em relação a distribuição de banda ?

## 1.2 - Revisão literária

A busca por técnicas que mitiguem a ICI vem sendo alvo de pesquisas constantes pelo meio acadêmico e 
entidades. Então, cabe, nesse momento, fazermos uma breve apresentação de trabalhos que foram direcionadas
para ICIC em sistemas móveis.

Os autores (Omar Albeiro Trejo Narváez, íctor Fabián Miramá Pérez, 2018) expoem trabalhos focados em alcançar redes auto-organizáveis usando técnicas de aprendizado de máquina (aprendizagem por reforço, Q-Learning, aprendizagem por estatística, SVM, e classificação de padrões) juntamente às técnicas de reuso de frequência.
Efetua uma comparação das técnicas de reuso de frequência (FFR, SFR e EFFR), e explana como cada algoritmo de aprendizado de máquina pode ser aplicado em ICIC.
Por fim, recomendam criar um fluxograma do problema a ser atacado, definindo os parâmetros referentes a quantidade UEs, potências, células, etc.,  antes de escolher a técnica de aprendizado.

Em (Michael Brehm, Ravi Prakash, 2013), é proposto um algoritmo de redes auto-organizadas (SON), do inglês self-
organizing network,  capaz de realocar, dinamicamente e de forma pró-ativa, recursos para a eNodeB que por ventura detectar picos de tráfego, HotSpot, e que esteja em condições de esgotamento de recursos. Além disso, a solução garante a coordenação de interferência entre células ou setores LTE.
O treinamento e a decisão são feitos pela técnica Support Vector Regression (SVR).

O artigo de (Tingting Zhu, Nan Liu, Zhiwen Pan, Xiaohu You, 2015) propõe dois algoritmos que ativam e desativam 
as small células, sendo que no primeiro algoritmo é feito um revezamento de subframe das small celulas que 
causam séria interferência para evitar interferência grave a qualquer momento. 
O segundo algoritmo proposto é um esquema semi-estático liga/desliga de células pequenas, onde as células pequenas
são desligadas se causarem interferência séria ou caso não haja nenhum usuário associado a ela.

Partindo para o uso de aprendizado de máquina, (Chien-Hao Lee, Kuang-Hsun Lin and Hung-Yu Wei, 2019) propoem um
algoritmo de otimização de recurso por meio da técnica Q-Learning, tomando como parâmetro de aprendizado o sinal de referência da potência recebida, do inglês Reference Signal Received Power (RSRP), e um algoritmo de agrupamento de usuários. Além disso, estima os resultados do Q-Learning para casos gerais via rede neural profunda, do inglês Deep Neural Network (DNN).

Os autores (Mohamad Yassin, Mohamed A. AboulHassan, Samer Lahoud, Marc Ibrahim, Dany Mezher, Bernard Cousin, Essam A. Sourour 2017)
fazem uma análise comparativa  das técnicas  de ICIC usando esquemas FFR e SFR com fator de reuso 1 e 3 e 
com distribuições de UE uniformes e não uniformes.Pelas simulações é possível estudar o desempenho de cada técnica,
para vários parâmetros: eficiência espectral, eficiência energética, throughput médio por zona, 
índice de equidade de throughput e satisfação do UE.

O recente trabalho dos autores (Iago Diógenes do Rego, Vicente A. de Sousa Jr, 2021), o qual é tomado como base para este, aborda bem o tema dos HotSpots e os danos causados ao desempenho dos sistemas LTE. Então, o apresenta como proposta otimizar, por meio do algoritmo de aprendizado de máquina Q-Learning, as técnicas de ICIC, as quais são ajustadas de forma dinâmica conforme surgimento de hotspots.
Além disso, mostra via simulações e comparações de resultados que das técnicas ICIC empregadas a que apresentou melhor desempenho foi a Strict Frequency Reuse pois apresentou melhor compromisso entre desempenho e complexidade, permitindo uma maior flexibilidade na configuração dos parâmetros.

É perceptível que as técnicas ICIC é um assunto de muito interesse  pelas indústrias de telecomunicações e diversas 
estão sendo desensolvidas com foco no aperfeiçoamento das mesmas.
Entretanto, a literatura ainda é carente em trabalhos que abordem esquemas de aprendizado de máquina aplicados 
a cenários dinâmicos com presença de HotSpots. Sendo assim, este trabalho irá avaliar o comportamento do sistema LTE ao ser aplicado o algoritmo de 
aprendizado por reforço, MAB, na configuração dos parâmetros das técnicas ICIC  e comparar aos resultados do algoritmo Q-Learning do (Iago Diógenes do Rego, Vicente A. de Sousa Jr, 2021).

## 3.1 - Simulador de redes NS-3 

É de suma importancia o uso de softwares na simulação de sistemas de redes móveis, concepção de novas tecnologias ou aperfeiçoamento dos métodos já utilizados, além disso , por meio dos simuladores é possível prototipar e efetuar possíveis correções do projeto antes de uma implantação errada que poderá gerar altos custos ou até mesmo comprometer todo o planejamento.
Existem vários simuladores de redes móveis, mas para esse projeto, optou-se pelo NS-3, do inglês Network Simulator.
O ns-3 é um simulador de redes gratuito e open source bastante utilizado nas pesquisas pelo meio acadêmico, além disso possui uma comunidade ativa e colaboradora na solução de problemas e discussões de relatos ocorridos no sistema ou possíveis melhorias.
